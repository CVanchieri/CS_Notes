## OpenCV Notes
#### OpenCV is the huge open-source library for the computer vision, machine learning, and image processing and now it plays a major role in real-time operation which is very important in today’s systems
 - can process images and videos to identify objects, faces, or even handwriting of a human
 - when integrated with various libraries, such as Numpuy, python is capable of processing the OpenCV array structure for analysis
 - to identify image pattern and its various features we use vector space and perform mathematical operations on these features
 - designed with a main focus of real-time applications for computational efficiency

#### Some Applications of OpenCV
- face recognition
- Automated inspection and surveillance
- number of people – count (foot traffic in a mall, etc)
- Vehicle counting on highways along with their speeds
- Interactive art installations
- Anamoly (defect) detection in the manufacturing process (the odd - defective products)
- Street view image stitching
- Video/image search and retrieval
- Robot and driver-less car navigation and control
- object recognition
- Medical image analysis
- Movies – 3D structure from motion
- TV Channels advertisement recognition

#### Functionality of OpenCV
- Image/video I/O, processing, display (core, imgproc, highgui)
- Object/feature detection (objdetect, features2d, nonfree)
- Geometry-based monocular or stereo computer vision (calib3d, - stitching, videostab)
- Computational photography (photo, video, superres)
- Machine learning & clustering (ml, flann)
- CUDA acceleration (gpu)

```
### import library ### 
import cv2 
from google.colab.patches import cv2_imshow # use for collab
### load in the image ## 
# loads in BGR not RGB
# 3 mode options for loading, regular color, greyscale, without transparency
# -1, cv2.IMREAD_COLOR: loads a color image, no transparency, default
# 0, cv2.IM_READ_GRAYSCALE: laods image in grayscale mode 
# 1, cv2.IM_READ_UNCHANGED: laods image as such including alpha channel
# grayscale
img = cv2.imread('logo.jpg', 1) # read in the path to the file, jpg, png 
### display the image ### 
cv2_imshow(img) # run imshow on the file
# only necessary locally 
cv2.waitKey(0) # wait certain amount of time, 0 - will wait until key is pressed
cv2.destroyAllWindows() # closes all windows once done

```

```
### resize an image ###
img = cv2.imread('logo.jpg', 1) # read in the path to the file, jpg, png 
img = cv2.resize(img, (400, 400)) # resize by pixels
# img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5) # resize by fractions
### display the image ### 
cv2_imshow(img) # run imshow on the file
# only necessary locally 
cv2.waitKey(0) # wait certain amount of time, 0 - will wait until key is pressed
cv2.destroyAllWindows() # closes all windows once done
```

```
### rotate an image ###
img = cv2.imread('logo.jpg', 1) # read in the path to the file, jpg, png 
img = cv2.resize(img, (400, 400)) # resize by pixels
# img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5) # resize by fractions
img = cv2.rotate(img, cv2.cv2.ROTATE_90_COUNTERCLOCKWISE) # set how to rotate
### display the image ### 
cv2_imshow(img) # run imshow on the file
# only necessary locally 
cv2.waitKey(0) # wait certain amount of time, 0 - will wait until key is pressed
cv2.destroyAllWindows() # closes all windows once done
```

```
### Save a new image ###
img = cv2.imread('logo.jpg', 1) # read in the path to the file, jpg, png 
img = cv2.resize(img, (400, 400)) # resize by pixels
# img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5) # resize by fractions
img = cv2.rotate(img, cv2.cv2.ROTATE_90_COUNTERCLOCKWISE) # set how to rotate
cv2.imwrite('new_img.jpg', img)
### display the image ### 
cv2_imshow(img) # run imshow on the file
# only necessary locally 
cv2.waitKey(0) # wait certain amount of time, 0 - will wait until key is pressed
cv2.destroyAllWindows() # closes all windows once done
```

```
### How is an image represented ###
# openCV uses BGR, blue green red not RGB red green blue
img = cv2.imread('logo.jpg', -1) # read in the path to the file, jpg, png 
# images are represented in a numpy array 
print('image type:', type(img))
print('image shape: height, width, channel(color space)', img.shape) # height, width, channel(color space)
print('--- image array ---')
print(img) # show the image array 
```

```
import random 
### how to modify image color ###
# openCV uses BGR, blue green red not RGB red green blue
img = cv2.imread('logo.jpg', -1) # read in the path to the file, jpg, png 
print(img[257][45:400]) # first row, pixels between 45-400
for i in range(100): # loop through first 100 
  for j in range(img.shape[1]): # loop through the first row 
    img[i][j] = [random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)] # set random colors 
### display the image ### 
cv2_imshow(img) # run imshow on the file
img = cv2.resize(img, (400, 400)) # resize by pixels
# only necessary locally 
cv2.waitKey(0) # wait certain amount of time, 0 - will wait until key is pressed
cv2.destroyAllWindows() # closes all windows once done
```

```
import random 
### how to copy a piece of an image ###
# openCV uses BGR, blue green red not RGB red green blue
img = cv2.imread('logo.jpg', -1) # read in the path to the file, jpg, png 
# find the part of array that you want 
tag = img[300:500, 350:650] # numpy array slice, slice within slice, rows, columns 
img[100:300, 650:950] = tag # where to copy the slice too, must be same dimension
img[100:300, 50:350] = tag # where to copy the slice too, must be same dimension
### display the image ### 
cv2_imshow(img) # run imshow on the file
img = cv2.resize(img, (400, 400)) # resize by pixels
# only necessary locally 
cv2.waitKey(0) # wait certain amount of time, 0 - will wait until key is pressed
cv2.destroyAllWindows() # closes all windows once done
```
